```python

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import json
import csv
import time

# Assuming you already have a logged-in driver instance
# driver = webdriver.Edge()  # Your existing driver
options = Options()
#options.add_argument("--headless")  # No GUI = much faster
# options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")
# options.add_argument("--disable-images")  # Don't load images
# options.add_argument("--disable-javascript")  # Careful! May break some sites
# options.add_argument("--blink-settings=imagesEnabled=false")
# options.add_argument("--disable-extensions")
# options.add_argument("--disable-plugins")
# options.add_argument("--disable-popup-blocking")

driver = webdriver.Edge(service=service, options=options)

def scrape_all_tables(driver):
    try:
        print('Extracting all table HTML on page...')
        
        tables_html = driver.execute_script("""
            var tables = document.querySelectorAll('table');
            var tableHTMLs = [];
            for (var i = 0; i < tables.length; i++) {
                tableHTMLs.push(tables[i].outerHTML);
            }
            return tableHTMLs;
        """)
        
        print(f"Found {len(tables_html)} tables")
        
        table_data = []
        for i, table_html in enumerate(tables_html):
            try:
                soup = BeautifulSoup(table_html, 'html.parser')
                table = soup.find('table')
                
                rows = []
                for tr in table.find_all('tr'):
                    cells = [td.get_text(strip=True) for td in tr.find_all(['td', 'th'])]
                    if cells:
                        rows.append(cells)
                
                if rows:
                    table_data.append({
                        'table_index': i,
                        'rows': rows,
                        'html': table_html
                    })
                    
            except Exception as e:
                print(f"Error processing table {i}: {e}")
                continue
        
        return table_data
        
    except Exception as e:
        print(f"Error extracting tables: {e}")
        return []



try:

    
    driver.get("https://finance.sina.com.cn/realstock/company/sh000001/nc.shtml")
    print('a1')
    # Wait for page to load
    #WebDriverWait(driver, 10)
    #.until(
    #    EC.presence_of_element_located((By.CSS_SELECTOR, "input[data-automation-id='email']"))
    #)
    print('a')
    time.sleep(5)
   
    close_button = driver.find_element(By.CSS_SELECTOR, "a[class='snp-btn-close-new']")
    driver.implicitly_wait(5)  # 5 seconds max for any element search
    print('d')
    close_button.click()

# Set page load timeout
    driver.set_page_load_timeout(15)  # 15 seconds max for page load

# Set script timeout
    driver.set_script_timeout(10)
    print('d2')
    data = scrape_tables_fast(driver)
    print("Scraped data:", data)

finally:
    # Keep browser open for inspection, remove if you want it to close automatically
    input("Press Enter to close browser...")
    driver.quit()






# Usage
```
